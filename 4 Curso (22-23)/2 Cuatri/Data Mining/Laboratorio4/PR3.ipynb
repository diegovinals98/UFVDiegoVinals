```python

import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder

data = pd.read_csv('./Dataset_lab04/enfermeria.csv')
data
```

```python
# Seleccionar las columnas relevantes para el clustering
columns = ['parents_occupation', 'childs_nursery', 'family_form', 'number_children', 'housing_conditions', 'financial_standing', 'social_conditions', 'health_conditions', 'recommendation']
# Seleccionar las características relevantes para el clustering

X = data[columns]

# Realizar el clustering jerárquico aglomerativo con diferentes configuraciones
methods = ['single', 'complete', 'average', 'ward']  # Diferentes métodos de enlace
distances = ['euclidean', 'manhattan', 'cosine']  # Diferentes tipos de distancias

for method in methods:
    for distance in distances:
        # Realizar el clustering jerárquico aglomerativo
        Z = linkage(X, method=method, metric=distance)

        # Generar el dendrograma
        plt.figure(figsize=(10, 6))
        dendrogram(Z, leaf_rotation=90, leaf_font_size=8)
        plt.title(f'Dendrograma - Método: {method.capitalize()}, Distancia: {distance.capitalize()}')
        plt.xlabel('Candidatos')
        plt.ylabel('Distancia')
        plt.show()
```
