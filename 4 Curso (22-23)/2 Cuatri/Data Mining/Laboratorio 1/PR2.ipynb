## Parte 1

### Vamos a relacionar en un gráfico el nivel de estudios con el saldo del cliente. Para ello agruparemos en tres grupos a los clientes que tienen deudas, a aquellos que tienen unos ahorros normales y a aquellos que disponen de suficientes ahorros.
#### Usaremos un diagrama de barras para cada grupo de saldo y, después, interpretaremos los resultados.

```python
# Leemos el csv banco
banco = pd.read_csv('./DatasetsLab1/banco ufv.csv', encoding='latin-1', sep=';')
# filtramos solo las columnas que nos interesan
bancofiltrado = banco[['age', 'education', 'balance', 'duration']]
bancofiltrado.head(20)
```

```python
# Añadimos una columna Rango vacia
#bancofiltrado["Rango"] = ""
bancofiltrado_copy = bancofiltrado.copy()
bancofiltrado_copy.loc[:, "Rango"] = ""
bancofiltrado_copy
```

### Rellenamos la columna Rango dependiendo de los valores de balance del dataset

```python
i = 0
for items in bancofiltrado_copy["balance"].values:
    # Si el balance es menor que 0, ponemos en la columna Rango "Deuda"
    if items < 0:
        bancofiltrado_copy.loc[i,"Rango"] = "Deuda"
        i = i + 1
    # Si el balance es menor que mayor que 0 y, mayor que 10.000, ponemos en la columna Rango "Positivo"       
    if items >= 0 and items < 10000:
        bancofiltrado_copy.loc[i,"Rango"] = "Positivo"
        i = i + 1
    # Si el balance es mayor que 10.000, ponemos en la columna Rango "Inversion "
    if items > 10000: 
        bancofiltrado_copy.loc[i,"Rango"] = "Inversion"
        i = i + 1
```

### Agrupamos por Rango y por tipo de Educación

```python
porRango = bancofiltrado_copy.groupby("Rango")["education"].value_counts()
# Cambiamos la columna index por "Frecuencia"
porRango = porRango.rename("Frecuencia")
# Lo convertimos en dataSet de la libreria pandas
df = porRango.reset_index()
df
```

### Dibujamos gráficas para cada tipo de rango

#### En este caso, el Rango es deuda

```python
# para Rango Deuda
# filtramos solo los datos que contengan Deuda en la columna Rango
eduacionDeuda = df.loc[df['Rango'] == "Deuda"]

# Dibujamos la grafica
plt.bar(eduacionDeuda["education"],eduacionDeuda["Frecuencia"])
# Le ponemos titulo a la grafica
plt.title('Saldo negativo (estan en deuda)')
# Ponemos lo que representa el eje X
plt.xlabel('Grado de Estudios')
# Ponemos lo que representa el eje Y
plt.ylabel('Numero de personas')
# Lo mostramos
plt.show()
```

Podemos observar que el mayor número de personas que están en deuda son las personas con grado secundario, es decir, personas que no tienen estudios universitarios. Estas personas no tienen título, por lo que tienen menos cantidad de ingresos y, por lo tanto, tienen que pedir más dinero al banco y endeudarse. Las personas con grado universitario tienen tendencia a tener mayor estabilidad económica y, por tanto, tienen menos deudas.

#### Rango: Positivo

```python
# Filtramos el dataset para cuando Rango es Posotivo
eduacionPositivo = df.loc[df['Rango'] == "Positivo"]
# Dibujamos el diagrama
plt.bar(eduacionPositivo["education"],eduacionPositivo["Frecuencia"])
# Le ponemos titulo al diagrama
plt.title('Saldo Postivo')
# Le ponemos el significado de la variable del eje X
plt.xlabel('Grado de Estudios')
# Le ponemos el significado de la variable del eje Y
plt.ylabel('Numero de personas')
# Mostramos el diagrama
plt.show()
```

Estas personas están más repartidas, podemos decir que son personas a las que, independientemente del nivel de estudios que tengan, saben manejar su dinero para no tener deudas con el banco. Cuanto mayor sea el grado de estudios, mayor será su capacidad para gestionar su dinero y, por tanto, tener unos ahorros normales (menos de 10.000 euros).

#### Rango: Inversion

```python
# Filtramos el dataset para los valores de rango igual a Inversion
eduacionInversion = df.loc[df['Rango'] == "Inversion"]
# Dibujamos los datos en un diagrama de barras
plt.bar(eduacionInversion["education"],eduacionInversion["Frecuencia"])
# Le ponemos titulo a la grafica
plt.title('Saldo Inversion')
# Ponemos el significado de la variable del eje X
plt.xlabel('Grado de Estudios')
# Ponemos el significado de la variable del eje Y
plt.ylabel('Nunero de personas')
# Mostramos la grafica
plt.show()
```

Podemos ver que los clientes con grados universitarios son los clientes con más ahorro y capacidad para ofrecerles paquetes de inversión, especialmente aquellos con más de 10.000 euros.

Conclusión final: El grupo de grados secundarios es el que destaca más. A pesar de no tener títulos universitarios, se distribuyen más o menos por igual en las tres categorías. Si no se observan los datos al principio, se podría pensar que ninguna persona de este grupo está en la categoría de inversión, ya que es poco común que alguien sin estudios tenga suficiente dinero para que un banco decida invertir con ellos. Sin embargo, es una clase media muy estable, con capacidad de invertir y mantener unos ahorros estables, a pesar de no tener un título universitario.

## Parte 2

### Haremos un gráfico donde se pueda ver cómo se distribuye cada clase en función del grupo de edad de los clientes.

```python
# Creamos una columna Rango de edad vacia
bancofiltrado_copy["RangoEdad"] = ""
```

### Rellenamos la columna RangoEdad según los valores de edad de los clientes.

```python
i = 0
for items in bancofiltrado_copy["age"].values:
    # si la edad es menor de 30 ponemos alumnos en la columna RangoEdad
    if items < 30:
        bancofiltrado_copy.loc[i,"RangoEdad"] = "Alumnos"
        i = i + 1
    if items >= 30 and items < 45:
        # si la edad es mayor de 30 y mayor de 45, ponemos Trabajadores en la columna RangoEdad
        bancofiltrado_copy.loc[i,"RangoEdad"] = "Trabajadores"
        i = i + 1
    if items >= 45 and items < 65:
         # si la edad es mayor de 45 y mayor de 65, ponemos Trabajadores Veteranos en la columna RangoEdad
        bancofiltrado_copy.loc[i,"RangoEdad"] = "Trabajadores Veteranos"
        i = i + 1
    if items >= 65: 
        # si la edad es mayor de 65, ponemos No Vinculados en la columna RangoEdad
        bancofiltrado_copy.loc[i,"RangoEdad"] = "No vinculados"
        i = i + 1
        
bancofiltrado.head(20)
```

### Contamos cuántos trabajadores, alumnos, trabajadores veteranos y no vinculados hay en el dataset.

```python
conteo = bancofiltrado_copy['RangoEdad'].value_counts()
# Lo convertimos en dataset
conteo.reset_index()
```

### Mostramos una gráfica con los datos.

```python
conteo.plot.pie(labels=conteo.index)
```

Más del 75% de los clientes son trabajadores, ya sean veteranos o no, y a su vez, más de la mitad de todos los trabajadores no son veteranos. En cambio, hay pocos estudiantes, lo que sugiere que este banco no es la mejor opción para estudiantes, ya que no ofrece las mismas ventajas que otros bancos podrían brindar, como menores comisiones por ser estudiante. Este banco parece ser más adecuado para personas adultas que trabajan, que ganan dinero y desean depositarlo en el banco para ahorrar mientras buscan comprar una casa, un coche u otras cosas similares.

En resumen, podemos decir que este banco está enfocado principalmente en personas activas en el mercado laboral, ya que más del 75% de sus clientes son trabajadores.

## Parte 3

### Queremos saber cómo se distribuyen y cuáles son las frecuencias en relación con los días que el cliente ha estado en la empresa. Para ello, transformaremos los días a años y lo representaremos gráficamente.
#### Para representarlo hemos pensado en un diagrama de dispersión.

### Creamos una nueva columna llamada "años", que representa el tiempo que los clientes han estado en el banco en años.

```python
bancofiltrado_copy["años"] = bancofiltrado_copy["duration"] / 365.25
bancofiltrado_copy
```

```python
# Años es la columna "años" del dataset
años = bancofiltrado_copy["años"]
# Edad es la columna "age" del dataset
edad = bancofiltrado_copy["age"]
```

### dibujamos los datos

```python
# dibujamos un diagrama scatter
plt.scatter(edad, años, color="blue", s=4)
# Ponemos titulo al eje x
plt.xlabel("Edad")
# Ponemos titulo al eje Y
plt.ylabel("Años")
# Ponemos titulo al diagrama
plt.title("Diagrama de dispersión de edad vs años")
# Mostramos el diagrama
plt.show()
```

Como mencionamos anteriormente, la mayoría de los clientes del banco son personas activas en el mercado laboral, con edades entre 25 y 60 años, lo cual se refleja en este diagrama. Es decir, podemos observar que el diagrama se mantiene estable en este rango de edades, sin embargo, a partir de los 60 años, podemos ver una disminución en la permanencia de los clientes en el banco, mientras que a partir de los 18 años comienza a aumentar.

Podemos observar que la duración más común de los clientes en el banco es de entre 3 y 5 años. También se aprecia que hay muchos clientes nuevos de todas las edades, lo que sugiere que el banco está obteniendo buenos resultados. Por otro lado, hay clientes que llevan casi 14 años en el banco, lo que nos permite inferir que la institución ha estado funcionando durante al menos 14 años (pocos años para ser un banco). Al haber tanta cantidad de nuevos clientes y tan pocos clientes con muchos años de permanencia, podemos decir que el banco es relativamente nuevo, con pocos años de funcionamiento.
