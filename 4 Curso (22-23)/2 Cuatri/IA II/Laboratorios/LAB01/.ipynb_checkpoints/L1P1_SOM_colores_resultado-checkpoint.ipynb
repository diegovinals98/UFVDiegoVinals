## SOM Setup
#### Variables definidas por el alumno

```python
# Inicializa tamaño del mapa de Kohonen, número de iteraciones y learning rate
# Inicializa normalizar_datos dependiendo de si tienes que normalizar los datos o no
lado_mapa = 30
periodo = 30000
learning_rate = 0.1
normalizar_datos = True
```

#### A partir de este punto solo hay cálculos. No se introducen más valores "a mano"

```python
# Establece el numero de entradas del mapa y el número de datos que se van a usar para entrenar. 
# Utiliza una función que obtenga automáticamente los valores a partir del Dataset.
num_entradas = datos.shape[0] # 3 entradas
num_datos = datos.shape[1] # 100 colores

# Calcula el vecindario inicial. Debe ser la mitad del lado del mapa de Kohonen
vecindario = lado_mapa/2

# Normaliza los datos si fuese necesario dividiendo cada dato por el máximo en la matriz
if normalizar_datos:
            datos = datos / datos.max()
    
# Crea una matriz de pesos con valores random entre 0 y 1. Usa la función random.random de la librería NumPy
matriz_pesos = np.random.random((lado_mapa,lado_mapa,num_entradas))
#pintar_mapa(matriz_pesos, periodo)
#aleatorio = datos[:, np.random.randint(0, num_datos)].reshape(np.array([num_entradas, 1]))
#aleatorio
```

#### Funciones para entrenar/clasificar

```python
# Función para encontrar la BMU
"""
   Encuentra la BMU para un patrón de entrada.
   Entradas: (patrón_de_entrada, matriz_de_pesos, número_de_entradas)
   Salidas:  (bmu, bmu_idx) tupla donde
               bmu: vector de pesos de la neurona ganadora
               bum_idx: coordenadas de la neurona ganadora
"""
def dist_v(vector1, vector2):
    # Calcula la distancia entre dos vectores
    return np.sum((vector1 - vector2) ** 2)

def calcular_bmu(patron_entrada, m_pesos, num_entradas):
    #Inicializamos las coordenadas de la bmu
    bmu_coord = np.array([0,0])
    #Le damos a la distancia_bmu un valor elevado al inicializarlo
    distancia_bmu = np.inf
    for i in range(m_pesos.shape[0]): #lado_mapa(1):
        for j in range (m_pesos.shape[1]): #lado_mapa(2)
            #El vector de pesos actualizado tiene forma de (3(num_entradas),1)
            v_pesos = m_pesos[i, j, :].reshape(num_entradas, 1)
            #Calculamos la distancia entre el patron de entrada y el vector de pesos
            dist = dist_v(patron_entrada, v_pesos)
            #Si la distancia del patron de entrada al vector de pesos
            #es menor al anterior registrado en distancia_bmu...
            if dist < distancia_bmu:
                distancia_bmu = dist
                #actualizamos las coordenadas de la neurona "hasta el momento" ganadora
                bmu_coord = np.array([i,j])
                bmu = m_pesos[bmu_coord[0],bmu_coord[1], :].reshape(num_entradas,1)
    return(bmu,bmu_coord)
            
          
```

```python
# Función para calcular el descenso del coeficiente de aprendizaje (eta)
"""
   Calcula el Learning Rate (eta) que corresponde a la i-ésima presentación.
   Entradas: (learning_rate_inicial, iteracion, período)
   Salidas:  learning_rate para la iteración i

"""
def variacion_learning_rate(lr_inicial, i, n_iteraciones):
    return lr_inicial * np.exp(-i / n_iteraciones)
```

```python
# Función para calcular el descenso del vecindario (v)
"""
   Calcula el vecindario  (v) que corresponde a la i-ésima presentación.
   Entradas: (vecindario_inicial, iteracion, período)
   Salidas:  lvecindario para la iteración i

"""
def variacion_vecindario(vecindario_inicial, i, periodo):
    return vecindario_inicial * np.exp(-i / periodo) + 1
```

```python
# Función para calcular el descenso del coeficiente de aprendizaje (eta) en función de la distancia a la BMU
"""
   Calcula la amortiguación de eta en función de la distancia en el mapa entre una neurona y la BMU.
   Entradas: (distancia_BMU, vecindario_actual)
   Salidas:  amortiguación para la iteración

"""
def decay(distancia_BMU, vecindario_actual):
    return np.exp(-distancia_BMU**2 / (2*vecindario_actual**2))
```

#### Funciones para dibujar la salida de la red

```python
# Función para pintar una matriz de valores como colores RGB
def pintar_mapa(matriz_valores,i):
    fig = plt.figure()
    
    # Establece ejes
    ax = fig.add_subplot(111, aspect='equal')
    ax.set_xlim((0, matriz_pesos.shape[0]+1))
    ax.set_ylim((0, matriz_pesos.shape[1]+1))
    ax.set_title('lr = {}, vecindario = {}, periodo = {}'.format(learning_rate, vecindario, i))

    # Dibuja los rectángulos de color RGB
    for x in range(1, matriz_valores.shape[0] + 1):
        for y in range(1, matriz_valores.shape[1] + 1):
            ax.add_patch(patches.Rectangle((x-0.5, y-0.5), 1, 1,
                         facecolor=matriz_valores[x-1,y-1,:],
                         edgecolor='none'))
    plt.show()
```

## SOM Entrenamiento

```python
def entrenamiento(datos, periodo):
    pintar_mapa(matriz_pesos, 0)
    for x in range(periodo):
        # Seleccionar una entrada aleatoria del conjunto de entrenamiento
        aleatorio = datos[:, np.random.randint(0, num_datos)].reshape(num_entradas, 1)
        # Calcular la neurona ganadora (unidad más cercana) para la entrada aleatoria
        bmu, bmu_coord = calcular_bmu(aleatorio, matriz_pesos, num_entradas)
        # Calculamos la distancia la bmu
        dist_bmu = dist_v(aleatorio, bmu)
        # actualizamos el descenso del vecindario
        vecindario_actualizado = variacion_vecindario(vecindario, x, periodo)
        # Actualizamos el learning rate
        learning_rate_actualizado = variacion_learning_rate(learning_rate, x, periodo)
        # Actualizar los pesos de la neurona ganadora y sus vecinos cercanos
        for i in range(matriz_pesos.shape[0]):
            for j in range(matriz_pesos.shape[1]):
                vector_pesos = matriz_pesos[i, j, :].reshape(num_entradas, 1)
                # Calculamos la distancia de las neuronas a la bmu
                dist_resto = np.sum((np.array([i, j]) - bmu_coord) ** 2)
                if dist_resto <= vecindario_actualizado ** 2:
                    # Si las neuronas están dentro del vecindario, actualizaremos la amortiguación y el peso
                    amortiguacion = decay(dist_resto**0.5, vecindario_actualizado)
                    variacion = vector_pesos + (amortiguacion * learning_rate_actualizado * (aleatorio - vector_pesos))
                    matriz_pesos[i, j, :] = variacion.reshape(1, 3)
        # cada 100 iteraciones hacemos un print de la matriz de pesos
        if (x % 100 == 0):
            print(x)
```

```python
entrenamiento(datos,periodo)
pintar_mapa(matriz_pesos, periodo)
```

## SOM Clasificación

```python
# Clasifica los patrones de entrenamiento con la matriz de pesos recién entrenada
def clasificacion(datos, matriz_pesos):
    #Definimos el numero de entradas
    num_entradas = datos.shape[0]
    #Definimos el tamaño de los datos
    num_datos = datos.shape[1]
    #Definimos el lado del mapa
    lado_mapa = matriz_pesos.shape[0]

    #Creamos array inicializado a 0s para guardar los valores de clasificacion
    mapa_clasificacion = np.zeros((lado_mapa, lado_mapa, num_entradas))
    #Creamos array inicializado a 0s para guardar los valores de activacion
    mapa_activaciones = np.zeros((lado_mapa, lado_mapa))
    #Creamos array inicializado a 0s para guardar los valores de distancias
    mapa_distancias = np.zeros((lado_mapa, lado_mapa))

    #Recorremos los datos de entrenamiento
    for i in range(num_datos):
        #Obtenemos el patrón de entrada
        patron_entrada = datos[:, i].reshape(num_entradas, 1)
        #Calculamos la BMU para el patrón de entrada
        bmu, bmu_coord = calcular_bmu(patron_entrada, matriz_pesos, num_entradas)
        #Imprimimos las coordenadas de la BMU
        print("Coordenadas BMU para patrón {}: ({},{})".format(i+1, bmu_coord[0], bmu_coord[1]))
        #Actualizamos el mapa de clasificación con el patrón de entrada
        mapa_clasificacion[bmu_coord[0], bmu_coord[1], :] = patron_entrada[:, 0]
        #Actualizamos el mapa de activaciones
        mapa_activaciones[bmu_coord[0], bmu_coord[1]] += 1
        #Actualizamos el mapa de distancias
        mapa_distancias[bmu_coord[0], bmu_coord[1]] += np.linalg.norm(patron_entrada - matriz_pesos[bmu_coord[0], bmu_coord[1], :])
    #Calculamos la distancia media y el error de cuantificacion
    distancia_total = 0
    for i in range(lado_mapa):
        for j in range(lado_mapa):
            #Calculamos la distancia media para cada neurona
            if mapa_activaciones[i, j] > 0:
                distancia_total += mapa_distancias[i, j] / mapa_activaciones[i, j]

    error_cuantificacion = distancia_total / (lado_mapa * lado_mapa)


    #Establecemos los valores NaN para las neuronas que no se han activado
    mapa_distancias[mapa_activaciones == 0] = np.nan
    #Calculamos la distancia media
    distancia_media = np.nanmean(mapa_distancias) / num_entradas
    print("Error cuantificacion: {}".format(error_cuantificacion))

    return (mapa_clasificacion, mapa_activaciones, mapa_distancias)

def pintar_resultados_clasificacion (mapa_clasificacion, mapa_activaciones, mapa_distancias):
    
    #3. Dibuja los mapas de clasificación (colores RGB), activaciones y distancias, y las clases
    pintar_mapa(mapa_clasificacion,periodo)

    #Pintamos el histograma del mapa de activación
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    x_data, y_data = np.meshgrid( np.arange(mapa_activaciones.shape[1]), np.arange(mapa_activaciones.shape[0]) )
    x_data = x_data.flatten()
    y_data = y_data.flatten()
    z_data = mapa_activaciones.flatten()
    ax.bar3d( x_data,
              y_data,
              np.zeros(len(z_data)),
              1, 1, z_data )
    plt.show()

    print(mapa_distancias)
    

(mapa_clasificacion, mapa_activaciones, mapa_distancias) = clasificacion(datos, matriz_pesos)
pintar_resultados_clasificacion(mapa_clasificacion, mapa_activaciones, mapa_distancias)
```
