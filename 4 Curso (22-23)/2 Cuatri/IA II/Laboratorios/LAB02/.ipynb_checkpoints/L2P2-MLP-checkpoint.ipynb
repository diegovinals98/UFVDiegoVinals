```python
import numpy as np
import keras as ke
from keras.models import Sequential
from keras.layers.core import Dense
import tensorflow as tf
import pandas as pd
```

Modelo RELU

```python
def modeloRELU (nuneroNuronas, epoch):
        
        # Creamos el modelo
        model = Sequential()
        # a침adimos una capa oculta, con hidden como numero de nueronas. Usa como funcoin de activacion relu
        model.add(Dense(nuneroNuronas, input_dim=2, activation='relu'))
        # Craemos una capa de salida cn una neurona, con funcion de activacion sigmoide
        model.add(Dense(1, activation='sigmoid'))
        # Compilamos el modelo con el optimizador adam, el error cuadratico medio 
        model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
        # ajuste del modelo con las entradas y salidas.
        model.fit(entradas, salidas, epochs=epoch, batch_size=4)
        # Sacamos el error y la precision del modelo con las entradas y salidas.
        #loss, accuracy = model.evaluate(entradas, salidas)
        loss = model.evaluate(entradas,salidas )
        print("Funcion de activacion: RELU")
        print(f"Neuronas: {nuneroNuronas}")
        print(f"Epoch: {epoch}")
        print('Loss:', loss[0])
        print('Accuracy:', loss[1])    
        # Sacamos la salida de la red neuronal para cada entrada
        salidas_red = model.predict(entradas)
        print("Predicciones:")
        print(f"{salidas_red}  + \n")
       
```

Modelo SIGMOIDE

```python

def modeloSIGMOIDE (nuneroNuronas, epoch):
        # Creamos el modelo
        model = Sequential()
        # a침adimos una capa oculta, con hidden como numero de nueronas. Usa como funcoin de activacion relu
        model.add(Dense(nuneroNuronas, input_dim=2, activation='sigmoid'))
        # Creamos una capa de salida cn una neurona, con funcion de activacion sigmoide
        model.add(Dense(1, activation='sigmoid'))
        # Compilamos el modelo con el optimizador adam, el error cuadratico medio 
        model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
        # ajuste del modelo con las entradas y salidas.
        history = model.fit(entradas, salidas, epochs=epoch, batch_size=4)
        # Sacamos el error y la precision del modelo con las entradas y salidas.
        #loss, accuracy = model.evaluate(entradas, salidas)
        loss = model.evaluate(entradas,salidas )
        print("Funcion de activacion: Sigmoide")
        print(f"Neuronas: {nuneroNuronas}")
        print(f"Epoch: {epoch}")
        print('Loss:', loss[0])
        print('Accuracy:', loss[1])    
        # Sacamos la salida de la red neuronal para cada entrada
        salidas_red = model.predict(entradas)
        print("Predicciones:")
        print(f"{salidas_red}  + \n")
        
```

Definimos la entrada y la salida esperada

```python

# Definimos las entradas de XOR
entradas = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
# las cuatro salidas esperadas en el mismo orden
salidas = np.array([[0],[1],[1],[0]], "float32")
```

Probamos con 2, 3 y 4 neuronas

```python
# Probamos con 2 neuronas
print("--------------------------- MODELO 1: 2 NUERONAS ---------------------------")
modeloSIGMOIDE(2,300)
modeloRELU(2,300)

# Probamos con 3 nueronas
print("--------------------------- MODELO 2: 3 NUERONAS ---------------------------")
modeloSIGMOIDE(3,600)
modeloRELU(3,600)

# Probamos con 4 nueronas
print("--------------------------- MODELO 3: 4 NUERONAS ---------------------------")
modeloSIGMOIDE(4,900)
modeloRELU(4,900)


```
