```python
import matplotlib.pyplot as plt
import pandas as pd
import random
import numpy as np
import copy

# Creamos el dataframe de las distancias
df_distancias = pd.read_excel("Datos_TSPincompleto.xlsx", sheet_name="Distancias", index_col = 0)
df_distancias = df_distancias.iloc[7:, :]
df_distancias.columns = df_distancias.iloc[0]
df_distancias = df_distancias[1:]

# Creamos el dataframe de las poblaciones
poblacion = pd.read_excel("Datos_TSPincompleto.xlsx", sheet_name="Población", index_col=0)

# Elimina las primeras 9 columnas del dataframe
df_poblacion = poblacion.iloc[3:, :]
df_poblacion.columns = df_poblacion.iloc[0]
df_poblacion = df_poblacion[1:]
```

```python
# Creamos un diccionario para las distancias que asocie cada nombre de columna con un número del 1 al L
# Asociamos cada municipio con un número del 1 al L
nombres_columnas = df_distancias.columns.tolist()
dicc_distancias = {nombre: numero+1 for numero, nombre in enumerate(nombres_columnas)}
df_distancias.rename(index=dicc_distancias, inplace=True)
df_distancias.rename(columns=dicc_distancias, inplace=True)
```

```python
# Creamos un diccionario para las distancias que asocie cada nombre de columna con un número del 1 al L
# Asociamos cada municipio con un número del 1 al L
nombres_columnas = df_poblacion.index.tolist()
dicc_poblacion = {nombre: numero+1 for numero, nombre in enumerate(nombres_columnas)}
df_poblacion.rename(index=dicc_poblacion, inplace=True)
```

```python
# Definimos las variables 
L = len(nombres_columnas) #Numero total de estaciones
NPOB = 100 #Tamaño de la poblacion
E = 8 #Numero de estaciones
NGEN = 50 #Numero de generaciones
pc = 0.4 #Probabilidad de cambio
pi = 0.3 #Probabilidad de intercambio
pnada = 1 - pc - pi #Probabilidad de no cambio
NRES = 10 #Numero de generaciones en las que se imprime el resumen
NSAMPLE = 10 #Numero de generaciones en las que se muestra la poblacion
alpha = 0.5
```

```python
def generar_cromosoma():
    cromosoma = []
    cromosoma = random.sample(range(1, L+1), E)
    return cromosoma

def generar_poblacion_inicial():
    poblacion = []
    numero_cromosoma = []
    for i in range(NPOB):
        numero_cromosoma.append(i+1)
        cromosoma = generar_cromosoma()
        poblacion.append(cromosoma)
    return poblacion, numero_cromosoma

poblacion_cromosoma = []
poblacion = []
num = 0

poblacion, numero_cromosoma = generar_poblacion_inicial()
for cromosoma in poblacion:
    poblacion_cromosoma.append((cromosoma))
#poblacion_cromosoma
```

```python
def calculate_fitness(df_distancias, df_poblacion, chromosome, alpha):
    total_column = df_poblacion['Total']
    dist = df_distancias
    p = total_column
    fitness = 0
    sum_p = 0
    sum_dist = 0
    if len(chromosome) != E:
        return fitness
    for i in range(E):
        sum_p = sum_p + p[chromosome[i]]
        if i < len(chromosome) - 1:
            sum_dist += dist[chromosome[i]][chromosome[i+1]]
        if i == E - 1:
            sum_dist += dist[chromosome[i]][chromosome[0]]
    fitness = alpha * sum_p - (1 - alpha) * sum_dist
    return fitness

# Generar una población de NPOB individuos aleatorios y calcular su fitness
poblacion_fitness = []
num = 0
for cromosoma in poblacion:
    fitness = calculate_fitness(df_distancias, df_poblacion, cromosoma, alpha)
    poblacion_fitness.append((num+1, cromosoma, fitness))
    num+=1


# Ordenar la población por su valor de fitness de mayor a menor
primera_poblacion = sorted(poblacion_fitness, key=lambda x: x[2], reverse=True)
#primera_poblacion
```

```python
def ind_mejores(poblacion_fitness_ordenada, porcentaje):
    num_mover = int(NPOB * porcentaje)
    # Obtener los mejores individuos de la generación padre
    mejores_padre = poblacion_fitness_ordenada[:num_mover]

    # Calcular la suma de los fitness de los mejores individuos
    suma_fitness_mejores = sum(fitness for _, _, fitness in mejores_padre)

    # Calcular la probabilidad de selección (Ps) para cada individuo
    tabla_auxiliar = []
    suma_ps = 0
    for numero_cromosoma,cromosoma, fitness in mejores_padre:
        ps = fitness / suma_fitness_mejores
        suma_ps += ps
        tabla_auxiliar.append((numero_cromosoma, cromosoma, fitness, ps, suma_ps))

    return tabla_auxiliar
```

```python
def seleccionar_cromosoma(tabla_auxiliar):
    r = random.random()  # Generar un número aleatorio entre 0 y 1
    for _, cromosoma, _, _,psa, in tabla_auxiliar:
        if r <= psa:
            return cromosoma
    return None
```

```python
def ruleta(pc, pi, pnada):
    opciones = ["pc", "pi", "pnada"]
    probabilidades = [pc, pi, pnada] #probabilidades de las opciones
    r = random.random() #generar un número aleatorio entre 0 y 1
    acumulado = 0
    for i in range(len(probabilidades)):
        acumulado += probabilidades[i] #acumular las probabilidades
        if r <= acumulado:
            return opciones[i] #retornar la opción seleccionada
```

```python
def intercambiar_cambiar_nada(tabla_padres, L, E, pc, pi, pnada):
  tabla_copia = copy.deepcopy(tabla_padres)
  cromosoma_seleccionado = seleccionar_cromosoma(tabla_copia)
  #print("Cromosoma inicial: " +str(cromosoma_seleccionado))
  for i in range(len(cromosoma_seleccionado)):

    #Seleccion de nuestra opcion por pc, pi y pnada método ruleta
    opcion = ruleta(pc, pi, pnada)

    #Opcion cambiar estación por otra no seleccionada
    if opcion == "pc":
      #Array para almacenar las estaciones que no se encuentren en el cromosoma
      estaciones_disponibles = []
      #Seleccionamos las estaciones/numeros que no esten en el cromosoma padre
      for j in range(1, L):
        if j not in cromosoma_seleccionado:
          #almacenamos las estaciones/numeros
          estaciones_disponibles.append(j)
      #Si E=L, no hay cambio. Si E<L entonces seleccionamos estaciones aleatorias
      if len(estaciones_disponibles) != 0:
        estacion_aleatoria = random.choice(estaciones_disponibles)
        # Reemplazar la estación elegida en el paso 1 por la estación elegida en el paso 2
        cromosoma_seleccionado[i] = estacion_aleatoria

    #Opcion cambiar orden estaciones
    elif opcion == "pi":
      #Si posicion bucle es la última, cambiar por la primera posicion
      if i == E-1:
        aux = cromosoma_seleccionado[i]
        cromosoma_seleccionado[i] = cromosoma_seleccionado[0]
        cromosoma_seleccionado[0] = aux
      #Para el resto de posiciones, cambiamos posicion actual por siguiente
      else:
        aux = cromosoma_seleccionado[i] 
        cromosoma_seleccionado[i] = cromosoma_seleccionado[i+1]
        cromosoma_seleccionado[i+1] = aux
  #print("Cromosoma final: " +str(cromosoma_seleccionado))
  #print("Cromosoma original: " + str(cromosoma))
  #print()
  return cromosoma_seleccionado
```

```python
def ind_peores(poblacion_fitness_ordenada, porcentaje):
    num_mover = int(NPOB * porcentaje)
    # Obtener los mejores individuos de la generación padre
    peores_padre = poblacion_fitness_ordenada[num_mover:]

    # Calcular la suma de los fitness de los mejores individuos
    suma_fitness_mejores = sum(fitness for _, _, fitness in peores_padre)

    # Calcular la probabilidad de selección (Ps) para cada individuo
    tabla_auxiliar = []
    suma_ps = 0
    for numero_cromosoma,cromosoma, fitness in peores_padre:
        ps = fitness / suma_fitness_mejores
        suma_ps += ps
        tabla_auxiliar.append((numero_cromosoma, cromosoma, fitness, ps, suma_ps))

    return tabla_auxiliar
```

```python
def generar_70(poblacion):
  """ Funcion que:
  Recibe la poblacion (anterior/primera)
  Devuelve el 70% de la nueva poblacion:
  - 20% (Mejores cromosomas de la poblacion padre)
  - 50% (Modificacion del 20% de la poblacion padre) """
  #Para una generacion inicial seleccionamos el 20% de los mejores
  tabla_padres = ind_mejores(poblacion, 0.2)
  # Creamos un array para almacenar el 50% de los valores
  poblacion_70 = []
  # Almacenamos el 20% 
  for _, cromosoma, _, _,_, in tabla_padres:
    poblacion_70.append(cromosoma)
  # Almacenamos el 50%
  numMover = int(NPOB*0.5)
  for i in range(numMover)  :
    cromosoma_cambios = intercambiar_cambiar_nada(tabla_padres, L, E, pc, pi, pnada)
    poblacion_70.append(cromosoma_cambios)
  return poblacion_70, tabla_padres
```

```python
def generar_30(poblacion):
  """ Funcion que:
  Recibe la poblacion (anterior/primera)
  Devuelve el 30% de la nueva poblacion"""
  tabla_peores = ind_peores(primera_poblacion,0.2)
  poblacion_30 = []
  #Almacenamos el 30%
  numMover30 = int(NPOB*0.3)
  for i in range(numMover30):
    cromosoma_cambios_tablaPeores = intercambiar_cambiar_nada(tabla_peores, L, E, pc, pi, pnada)
    poblacion_30.append(cromosoma_cambios_tablaPeores)
  return poblacion_30
```

```python
def generar_100(poblacion70, poblacion30):
  """Funcion que:
  Recibe:
  - Poblacion Nueva 70%
  - Poblacion Nueva 30%
  Devuelve:
  - Tabla(100% Poblacion Nueva y ordenado por fitness)"""
  poblacion100 = []
  poblacion70.extend(poblacion30)
  poblacion100 = copy.deepcopy(poblacion70)
  # Calculamos el fitness para cada cromosoma de la poblacion completa y lo ordenamos.

  poblacion100_fitness = []
  num = 0
  for cromosoma in poblacion100:
    fitness = calculate_fitness(df_distancias, df_poblacion, cromosoma, alpha)
    poblacion100_fitness.append((num+1, cromosoma, fitness))
    num+=1

  #poblacion100_fitness = calculate_fitness(poblacion100)
  # Ordenar la población por su valor de fitness de mayor a menor
  poblacion_100_fitness = sorted(poblacion100_fitness, key=lambda x: x[2], reverse=True)
  return poblacion_100_fitness
```

```python
def get_consensus_chromosome(population):
    # Obtener la longitud del cromosoma
    chromosome_length = len(population[0])
    
    # Crear una lista para almacenar el cromosoma consenso
    consensus_chromosome = []

    # Iterar sobre cada posición del cromosoma
    for i in range(chromosome_length):
        # Crear un diccionario para contar las ocurrencias de cada estación en esa posición
        station_counts = {}

        # Iterar sobre cada individuo de la población y contar las ocurrencias de las estaciones
        for individual in population:
            station = tuple(individual[1])  # Convertir la lista en una tupla
            if station in station_counts:
                station_counts[station] += 1
            else:
                station_counts[station] = 1

        # Obtener el número máximo de ocurrencias de una estación en esa posición
        max_count = max(station_counts.values())

        # Obtener todas las estaciones más comunes que tienen el máximo número de ocurrencias
        most_common_stations = [list(station) for station, count in station_counts.items() if count == max_count]

        # Seleccionar aleatoriamente una de las estaciones más comunes como el cromosoma consenso para esa posición
        selected_station = random.choice(most_common_stations)

        # Agregar la estación al cromosoma consenso
        consensus_chromosome.append(selected_station)

    return consensus_chromosome

```

```python
nueva_poblacion70, tabla_padres = generar_70(primera_poblacion)
nueva_poblacion30 = generar_30(primera_poblacion)
nueva_poblacion = generar_100(nueva_poblacion70, nueva_poblacion30)

tabla_padres_df = pd.DataFrame(tabla_padres, columns=["Numero","Cromosoma", "Fi", "Ps", "Psai"])
tabla_padres_df = tabla_padres_df[["Numero", "Fi", "Ps", "Psai"]]  # Seleccionar solo las columnas deseadas

# Calcular la suma de la columna "Fi"
suma_fi = tabla_padres_df["Fi"].sum()

# Configurar opciones de visualización
pd.set_option("display.max_columns", None)
pd.set_option("display.expand_frame_repr", False)

# Mostrar el DataFrame como una tabla con líneas
print(tabla_padres_df.to_string(index=False, line_width=None))
```

```python
print("Poblacion inicial: ", nueva_poblacion)
print()

#Almacenar valores de BESTF, BESTn y BEST% para graficar
BESTf = []
BESTn = []
BEST_porcentaje = []
consenso_fitness = []
# Crear una lista para almacenar las medias del fitness por generación
media_fitness_por_generacion = []

for i in range(NGEN):
    nueva_poblacion70, tabla_padres = generar_70(nueva_poblacion)
    nueva_poblacion30 = generar_30(nueva_poblacion)
    nueva_poblacion = generar_100(nueva_poblacion70, nueva_poblacion30)
    #print(nueva_poblacion)

    num_mover = int(NPOB * 0.2)
    sample = nueva_poblacion[:num_mover]
    media_fitness = np.mean([cromosoma[2] for cromosoma in sample])

    # Encontrar el mejor cromosoma y su fitness
    best_chromosome = max(nueva_poblacion, key=lambda x: x[2])
    best_fitness = best_chromosome[2]
    # Contar cuántos cromosomas tienen el mismo fitness que el mejor
    num_best = sum(1 for ind in nueva_poblacion if ind[2] == best_fitness)
    # Calcular el porcentaje de la población que tiene el mejor fitness
    percent_best = (num_best / len(nueva_poblacion)) * 100
    # Encontrar el cromosoma consenso y su fitness
    consensus_chromosome = get_consensus_chromosome(nueva_poblacion)
    consensus_fitness = calculate_fitness(df_distancias, df_poblacion, consensus_chromosome[0], alpha)
        

    media_fitness_por_generacion.append(media_fitness)
    BESTf.append(best_fitness)
    BESTn.append(num_best)
    BEST_porcentaje.append(percent_best)
    consenso_fitness.append(consensus_fitness)
    
    if (i+1) % NRES == 0:        

        print("GENERACIÓN", i+1)
        print("\tMejor cromosoma:", best_chromosome[1])
        print("\tFitness (BESTf):", best_fitness)
        print("\tCuantos hay (BESTn):", num_best)
        print("\tEn % (BEST%):", percent_best)
        print("\tCromosoma consenso:", consensus_chromosome[0])
        print("\tFitness (CONSf):", consensus_fitness)
        print()
    
    if (i+1) % NSAMPLE == 0:
      num_mover = int(NPOB * 0.5)
      sample = nueva_poblacion[:num_mover]  # Seleccionar el 20% con el mejor fitness
        
      print("MUESTREO DE LA GENERACIÓN", i+1)
      for ind in sample:
          print("\tCromosoma:", ind[1])
          print("\tFitness:", ind[2])
      print()
```

```python
# Crear una figura con una matriz de subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

generaciones = list(range(1, NGEN + 1))
# Gráfico BESTf
axs[0, 0].plot(generaciones, BESTf, label='BESTf', color='blue')
axs[0, 0].set_xlabel('Generación')
axs[0, 0].set_ylabel('BESTf', color='blue')
axs[0, 0].tick_params(axis='y', labelcolor='blue')
axs[0, 0].legend()
axs[0, 0].set_title(f'Evolución de BESTf (NPOB={NPOB}, alfa={alpha}, NGEN={NGEN})')

# Gráfico BESTn
axs[0, 1].plot(generaciones, BESTn, label='BESTn', color='red')
axs[0, 1].set_xlabel('Generación')
axs[0, 1].set_ylabel('BESTn', color='red')
axs[0, 1].tick_params(axis='y', labelcolor='red')
axs[0, 1].legend()
axs[0, 1].set_title(f'Evolución de BESTn (NPOB={NPOB}, alfa={alpha}, NGEN={NGEN})')

# Gráfico BEST%
axs[1, 0].plot(generaciones, BEST_porcentaje, label='BEST%', color='green')
axs[1, 0].set_xlabel('Generación')
axs[1, 0].set_ylabel('BEST%', color='green')
axs[1, 0].tick_params(axis='y', labelcolor='green')
axs[1, 0].legend()
axs[1, 0].set_title(f'Evolución de BEST% (NPOB={NPOB}, alfa={alpha}, NGEN={NGEN})')

# Gráfico consf
axs[1, 1].plot(generaciones, consenso_fitness, label='Consf', color='blue')
axs[1, 1].set_xlabel('Generación')
axs[1, 1].set_ylabel('CONSf', color='blue')
axs[1, 1].tick_params(axis='y', labelcolor='blue')
axs[1, 1].legend()
axs[1, 1].set_title(f'Evolución de CONSf (NPOB={NPOB}, alfa={alpha}, NGEN={NGEN})')

# Ajustar el espaciado entre los subplots
plt.tight_layout()

# Mostrar el collage de gráficos
plt.show()

```
