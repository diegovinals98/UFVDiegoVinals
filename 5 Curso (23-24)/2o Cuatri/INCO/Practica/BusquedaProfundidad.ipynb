```python
def es_valido(sudoku, fila, col, num):
    # Verificar si el número ya existe en la fila o columna
    for i in range(9):
        if sudoku[fila][i] == num or sudoku[i][col] == num:
            return False

    # Verificar si el número ya existe en el bloque de 3x3
    inicio_fila = fila - fila % 3
    inicio_col = col - col % 3
    for i in range(3):
        for j in range(3):
            if sudoku[inicio_fila + i][inicio_col + j] == num:
                return False
    return True  # Retorna True si el número puede ser colocado en la posición
```

# Funcion que resuelve el sudoku con Búsqueda en Profundidad

```python
def resolver_sudoku(sudoku, profundidad=0):
    # Encuentra la primera celda vacía en el sudoku
    vacia = encontrar_vacia(sudoku)
    if not vacia:
        return True  # Retorna True si no hay celdas vacías, lo que significa que el sudoku está resuelto

    fila, col = vacia  # Desempaqueta la posición de la celda vacía

    # Intenta colocar números del 1 al 9 en la celda vacía
    for num in range(1, 10):
        if es_valido(sudoku, fila, col, num):
            sudoku[fila][col] = num
            # Imprime la asignación actual con indentación basada en la profundidad de la recursión
            print(f"{'-' * profundidad}Asignar {num} en ({fila}, {col})")

            # Recursivamente intenta resolver el resto del sudoku
            if resolver_sudoku(sudoku, profundidad + 1):
                return True

            # Si no se encuentra solución, realiza backtrack
            sudoku[fila][col] = 0
            print(f"{'-' * profundidad}Retroceder en ({fila}, {col})")

    return False  # Retorna False si no se encuentra ninguna solución válida para esta rama
```

```python
def encontrar_vacia(sudoku):
    # Busca una celda vacía en el sudoku (representada por 0)
    for i in range(len(sudoku)):
        for j in range(len(sudoku[0])):
            if sudoku[i][j] == 0:
                return (i, j)  # Retorna la posición de la celda vacía
    return None  # Retorna None si no hay celdas vacías


def imprimir_sudoku(sudoku):
    for row in sudoku:
        print(" ".join(str(num) if num != 0 else '.' for num in row))
    print("\n" + "-" * 21 + "\n")
```
